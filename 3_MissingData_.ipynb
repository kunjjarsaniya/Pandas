# Handling Missing Data in Pandas

Welcome to this short tutorial on **handling missing data** using the `pandas` library.  
By the end of this lesson, you'll be able to:

* Detect where data is missing and quantify it.
* Decide when to **remove** or **fill** missing values.
* Apply common strategies like mean‚Äëimputation, constant filling, and row/column thresholding.

> **Prerequisites**: Basic familiarity with Python and pandas (creating a `DataFrame`, selecting columns, basic methods).

---

Let's get started!

```python
# Import the core libraries we'll use throughout the notebook
import numpy as np 
import pandas as pd 
```

## üìç Finding Missing Data

Below we explore pandas utilities that help you **identify** missing values.

**Finding Missing Data**

```python
# 1Ô∏è‚É£ Create a sample DataFrame that intentionally contains `NaN` values
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
# 2Ô∏è‚É£ Inspect the raw DataFrame
df
```

```python
# 3Ô∏è‚É£ Locate missing values: `isna()` shows True where data is missing
df.isna()
```

```python
# 4Ô∏è‚É£ Count missing values per column
df.isna().sum()
```

```python
# 5Ô∏è‚É£ Quickly check if **any** value is missing in each column
df.isna().any()
```

## üóëÔ∏è Removing Missing Data

Sometimes the simplest solution is to drop rows or columns that contain too many nulls.

**Removing Missing Data**

```python
# 6Ô∏è‚É£ Re‚Äëdisplay the original DataFrame before dropping rows
df
```

```python
# 7Ô∏è‚É£ Drop **any** rows that contain at least one missing value
df.dropna()
```

```python
# 8Ô∏è‚É£ The original `df` is unchanged because we didn't assign the result
#    (uncomment the next line to see the dropped version ‚û°Ô∏è df.dropna())
df
```

```python
# 9Ô∏è‚É£ Drop rows that have *more* than a threshold of missing values
#    (`thresh=1` keeps rows with at least 1 non‚Äënull value)
df.dropna(thresh=1)
```

## üß© Filling Missing Data

When dropping data isn't an option, we can *impute* ‚Äì i.e., plug reasonable values in place of the missing ones.

**Filling the missing Data**

```python
# üîü Check the DataFrame again before we start filling values
df
```

```python
# 1Ô∏è‚É£1Ô∏è‚É£ Replace every missing value with **0**
df.fillna(0)
```

```python
# 1Ô∏è‚É£2Ô∏è‚É£ Fill missing values with **column‚Äëspecific** replacements
values = {'A':0,'B':100,"C":300,'D':400}
df.fillna(value=values)
```

```python
# 1Ô∏è‚É£3Ô∏è‚É£ Inspect the DataFrame after custom filling
df
```

```python
# 1Ô∏è‚É£4Ô∏è‚É£ Fill numeric columns with their **mean** value (a common choice)
df.fillna(df.mean())
```
